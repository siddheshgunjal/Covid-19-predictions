# EWMA

```python
span = 4
alpha = 2/(span+1)
```

```python
df_confirmed.ewm(alpha=alpha, adjust=False, axis=1).mean()['2020-04-24']
```

```python
def cal_ewma(input_df):
    """Calculates the exp wighted moving average using the window size."""
    _df = input_df.ewm(alpha=alpha, adjust=False, axis=1).mean()['4/24/20']
    
    return _df
```

```python
cal_ewma(df_confirmed)
```

```python
all_pred_df = pd.DataFrame()

for df in [df_confirmed]:
    #_df = cal_moving_avg(df, window=2)
    ew_df = cal_ewma(df)
    #all_pred_df = pd.concat([all_pred_df, _df], axis=1)
    all_pred_df = pd.concat([all_pred_df, ew_df])
    
all_pred_df.columns =['Confirmed']    
#all_pred_df_ew.columns =['Confirmed', 'Recovered', 'Death']
```

```python
predictions = pd.DataFrame()

for n in range(0,188):
    model_1 = ExponentialSmoothing(endog=df_confirmed.iloc[n],trend='add').fit()
    fcast1 = model_1.forecast(1)
    fcast1 = round(fcast1, 1)
    predictions = pd.concat([predictions, fcast1])

predictions.columns = ['Confirmed']
predictions.set_index(df_confirmed.index, inplace=True)
predictions
```

```python
a = pd.DataFrame()
b = pd.DataFrame()

for n in range(0,188):
    model_1 = ExponentialSmoothing(endog=df_confirmed.iloc[n],trend='add').fit()
    model_2 = ExponentialSmoothing(endog=df_death.iloc[n],trend='add').fit()
    fcast1 = model_1.forecast(1)
    fcast2 = model_2.forecast(1)
    #print(fcast2)
    a = pd.concat([a, fcast1])
    b = pd.concat([b, fcast2])
    
#predictions = pd.merge(a, b)
#predictions.columns = ['Confirmed', 'Deaths']
#predictions.set_index(df_confirmed.index, inplace=True)
#predictions
```
